```python
import matplotlib.pyplot as plt
import numpy as np

# Models and datasets
models = ["Factorized Prior (GDN)", "Factorized Prior (ReLU)", "Scale Hyperprior"]
datasets = ["CelebA", "Imagenette"]

# SSIM values (replace with your actual data)
ssim_before = [
    [0.96, 0.89],  # Model A: [Dataset 1, Dataset 2] SSIM before
    [0.96, 0.89],  # Model B: [Dataset 1, Dataset 2] SSIM before
    [0.97, 0.91]   # Model C: [Dataset 1, Dataset 2] SSIM before
]

ssim_after = [
    [0.96, 0.89],  # Model A: [Dataset 1, Dataset 2] SSIM before
    [0.96, 0.89],  # Model B: [Dataset 1, Dataset 2] SSIM before
    [0.97, 0.91]   # Model C: [Dataset 1, Dataset 2] SSIM before
]

# Bar width and positions
bar_width = 0.2
x = np.arange(len(models))  # Label locations for models

# Add some space between the datasets
space_between_datasets = 0.42  # Adjust this value to control the space

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting before and after SSIM for each dataset
# Dataset 1 bars (CelebA)
ax.bar(x - bar_width / 2, [ssim_before[i][0] for i in range(len(models))], 
       bar_width, label='SSIM Before Attack - CelebA')
ax.bar(x + bar_width / 2, [ssim_after[i][0] for i in range(len(models))], 
       bar_width, label='SSIM After Attack - CelebA')

# Dataset 2 bars (Imagenette), with space between the datasets
ax.bar(x + space_between_datasets - bar_width / 2, [ssim_before[i][1] for i in range(len(models))], 
       bar_width, label='SSIM Before Attack - Imagenette')
ax.bar(x + space_between_datasets + bar_width / 2, [ssim_after[i][1] for i in range(len(models))], 
       bar_width, label='SSIM After Attack - Imagenette')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Models')
ax.set_ylabel('SSIM')
#ax.set_title('SSIM Before and After Attack for Each Model and Dataset')
ax.set_xticks(x + space_between_datasets / 2)
ax.set_xticklabels(models)
ax.legend()

# Display the SSIM values on top of the bars
for bars in ax.containers:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
